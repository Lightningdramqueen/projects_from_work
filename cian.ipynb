### Типы сделки

Посмотрим на общюю ситуцацию на рынке в разбивке по типу сделки:

```python
pd.DataFrame(df_merged['тип_сделки'].value_counts(normalize=True).mul(100).round(2))
```

```python
plt.figure(figsize=(10,7))

plt.subplot(2, 2, 1) 
ax1 = sns.countplot(data=df_merged, x='тип_сделки')
plt.title('Распределение на рынке недвижимости по типу сделки\n (абсолютные значения)', fontsize = 9)
# Добавление абсолютных значений на график
for p in ax1.patches:
    height = p.get_height()
    ax1.text(p.get_x() + p.get_width() / 2, height, f'{height}', ha='center', va='bottom')

# Создание пай чарта
plt.subplot(2, 2, 2)
deal_counts = df_merged['тип_сделки'].value_counts()
plt.pie(deal_counts, labels=deal_counts.index, autopct='%1.1f%%')
plt.title('Распределение на рынке недвижимости по типу сделки\n (относительные значения)', fontsize = 9)
plt.axis('equal')  # Сохранение равных пропорций для круга
```

В целом, тип сделки "Аренда" преобладает во всех регионах. Регионы, которые хотелось бы рассмотреть детальнее - нет.

**Больше городской недвижимости:** Общий анализ показывает, что большее количество сделок совершается с городской недвижимостью по сравнению с загородной. Это может быть связано с большим спросом на городскую недвижимость или с большим количеством доступной городской недвижимости на рынке.

Посмотрим детальнее на распредление по категории недвижимости (в разрезе типа сделки):

```python
df_merged.groupby(['категория_недвижимости', 'тип_сделки'])\
    .agg({'calculation1':'count'})\
    .rename(columns={'calculation1':'cnt'})\
    .reset_index()\
    .pivot_table(index=['тип_сделки'], columns='категория_недвижимости', values='cnt')
```

**Сюрпризов нет:** городская недвижимость чаще арендуется, в то время как загородная недвижимость продается и арендуется примерно с одинаковой частотой. Это может быть связано с тем, что городская недвижимость обычно включает в себя апартаменты и квартиры, офисы, которые чаще арендуются. В то время как загородная недвижимость, включающая в себя дома и участки, часто приобретается для собственности или инвестиций.
